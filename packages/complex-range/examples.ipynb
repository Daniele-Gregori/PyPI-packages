# Examples of complex-range 

Generate ranges of complex numbers — rectangular grids or linear (diagonal) sequences in the complex plane.

```python
from complex_range import complex_range, farey_sequence
```

## Rectangular Ranges

When the first argument is a plain number (not a list), `complex_range` generates a rectangular grid in the complex plane.

### From the origin

```python
complex_range(2+2j)
```

### Between two points

```python
complex_range(1+1j, 2+2j)
```

### With a complex step

The real part of the step controls the real axis spacing, the imaginary part controls the imaginary axis spacing.

```python
complex_range(0, 2+2j, 0.5+0.5j)
```

### Increment order

By default the imaginary component increments first (`increment_first='im'`). Use `'re'` to increment the real component first.

```python
# Default: imaginary first
complex_range(-1-1j, 1+1j, 1+1j, increment_first='im')
```

```python
# Real first
complex_range(-1-1j, 1+1j, 1+1j, increment_first='re')
```

### Farey subdivision

With `farey_range=True` and an integer step, intermediate points are placed according to the Farey sequence of that order, creating a finer grid.

```python
#wrong for fractional bounds
complex_range(0, .5+1.5j, [3,3], farey_range=True)
```

## Linear Ranges

When the first argument is a **list** (or tuple), `complex_range` generates collinear points along a line in the complex plane.

### Two-point diagonal

```python
complex_range([0, 2+2j])
```

### With a list step

```python
complex_range([0, 4+4j], [2, 2])
```

### With a complex step

```python
# should be deactivated
complex_range([0, 4+4j], 2+2j)
```

### Descending linear range

```python
complex_range([2+2j, 0], -1-1j)
```

## Notes

- If `z1` is a **list or tuple** → linear range (collinear points)
- Otherwise → rectangular range (grid)
- Default iteration order for rectangular ranges is *imaginary first*: for each real value, all imaginary values are enumerated before moving to the next real value
